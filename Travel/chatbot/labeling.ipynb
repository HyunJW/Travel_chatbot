```python
import pandas as pd
save = pd.read_csv('../data/수정데이터/맞춤법수정데이터.csv')
nouns = pd.read_csv('../data/수정데이터/명사들.csv')
```

```python
ch = save['checked']
```

```python
ch[0]
```

```python
ch = ch.dropna().reset_index(drop=True)
```

```python
ch.shape
```

```python
noun = nouns[nouns['label']!='O']
```

```python
noun['nouns'].tolist()
```

```python
from konlpy.tag import Komoran
komoran = Komoran()
```

```python
# 단어 시퀀스 생성
sequences = []
for i in range(len(ch)):
    keywords = komoran.nouns(ch[i])
    sequences.append(keywords)
```

```python
ch = ch.to_frame('text')
```

```python
df = pd.concat([ch, pd.DataFrame({'nouns':sequences})], axis=1)
df.tail()
```

```python
df.isnull().sum()
```

```python
df['nouns'][0][0] in noun['nouns'].values
```

```python
tags = []
for i in range(len(df)):
    temp = []
    for word in df.loc[i, 'nouns']:
        if word in noun['nouns'].values:
            tag = noun[noun['nouns'] == word]['label'].values[0]
        else:
            tag = 'O'
        temp.append((word, tag))
    tags.append(temp)
```

```python
tags
```

```python
labeling = pd.concat([df, pd.DataFrame({'match':tags})], axis=1)
labeling.tail()
```

```python
labeling.to_csv('../data/수정데이터/ner_train_data.csv', index=False)
```
